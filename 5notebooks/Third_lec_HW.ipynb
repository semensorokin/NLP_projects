Затем нам необходимо проделать эти операции и для твитов, тональность которых отмечена негативной оценкой

```python
f = open('mchsgov_all.txt').read()
mchsgov_articles = f.split('\n=============\n')
clear_mchsgov_art = []
mchsgov_words = []
for i in mchsgov_articles:
    words = parse_news(i)
    clear_mchsgov_art.append(' '.join(words))
    mchsgov_words.extend(words)
set_mchsgov_words = set(mchsgov_words)
print(len(mchsgov_words), len(set_mchsgov_words))
```
