```python
import pandas as pd
row_data = pd.read_excel('cloze_task_raw_data.xlsx')
print(row_data.shape)
row_data = row_data[row_data['shown'] != 'Введите первое слово']
print(row_data.shape)

def clear(x):
    flag=0
    if (len(x)==1) and (x not in 'ксявуоиа'):
        flag+=1
    for i in x:
        if i not in 'ячсмитьбюэждлорпавыфйцукенгшщзхъЯЧСМИТЬБЮФЫВАПРОЛДЖЭЪХЗЩЙЦШУГКНЕ':
             flag+=1
    if flag==0:
        return x
    else:
        return 0
row_data.answer = row_data.answer.astype(str)
row_data.answer = row_data.answer.apply(clear)
row_data = row_data[row_data['answer']!=0]
row_data.shape
```

```python
import fileinput


def extract_vectors(words, slim_model_path, big_model_path):
    '''
    Args:
      words(list) -- список слов, для которых нужны векторы
      slim_model_path(str) -- путь для сохранения извлечённых векторов в том же формате (без указания формата)
      big_model_path(str) -- путь к огромному файлу
    
    После работы функции список words будет пустым, если найдены все слова.
    '''
    with open(slim_model_path + '.vec', 'w') as file:
        for line in fileinput.FileInput(files=[big_model_path]):
            file.write(str(len(words)) + ' ' + line.split()[1] + '\n') # Слабое место, если не все слова найдены, то первая цифра в строчке не будет
            break                                                      # отображать реальное количество векторов в файле и он не прочитается генсимом.
        for line in fileinput.FileInput(files=[big_model_path]):       # В этом случае открываешь в Notepad++ или др. тектовом редакторе и правишь её
            vector = line.split()                                      # на реальное количество векторов в файле.
            if any([word == vector[0] for word in words]):
                file.write(' '.join(vector) + '\n')
                print(vector[0])
                del words[words.index(vector[0])]
                print('Words to find: {}'.format(len(words)))
            elif not words:
                break
                
extract_vectors(row_data.answer.unique().tolist(), 'Close_task_answers.vec', '/home/semen/Downloads/wiki.ru.vec')
#word_vectors_en_top = KeyedVectors.load_word2vec_format('wiki_top.en.vec')
```

```python
!ls
h = open('Close_task_answers.vec.vec', 'r').readlines()
word2ind = {}
vecs = []
for ind, j in enumerate(h[1:]):
    word2ind[j.split(' ')[0]] = ind
    vecs.append(j.split(' ')[1:])
```

```python
len(vecs), len(list(word2ind.items()))
```

```python
import numpy as np
from sklearn.cluster import KMeans
km = KMeans(n_clusters=40, init='k-means++', n_init=10, max_iter=500).fit_predict(np.array(vecs))
```

```python
ind2word = {j:i for i,j in word2ind.items()}
```

```python
clusters = []
for word_cluster in range(40):
    new_cluster = []
    for ind, i in enumerate(km):
        if i == word_cluster:
            new_cluster.append(ind2word[ind])
    clusters.append(new_cluster)
print(sum(len(i) for i in clusters))
    
```
